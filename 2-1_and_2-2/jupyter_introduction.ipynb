## マジックコマンド

### 先頭に % をつけられたマジックコマンドは一行を対象にする

````python
# 実行時間を測定
%time range(1000000)
````

````python
# 環境変数を表示
%env PATH
````

````python
# 返り値を取得することも可能
path = %env PATH
print(path.split(":")[0])
````

````python
# 変数一覧 %who および　詳細情報 %whos
%whos
````

````python
# マジックコマンドについてのヘルプを表示
%quickref
````

### 先頭に %% をつけられたマジックコマンドはセル全体を対象にする

* %%bash  
セル全体をシェルスクリプトとして実行

````python
%%bash
for i in `seq 1 10`
do
  echo $i
done
````

* %%time or %%timeit  
実行時間を計測

````python
%%time
sum = 0
for x in range(100000):
    sum += x
print(sum)
````

## グラフ描画ライブラリmatplotlib

````python
from matplotlib import pyplot as plt
````

````python
# 描画をmatplotlibの画面内に出力するためのマジックコマンド
%matplotlib inline
````

````python
plt.scatter([1,2,3,4,5],[2,3,4,5,6])
````

````python
import numpy as np
def spirograph(R=100, r=19, a=5):
    theta = np.linspace(0, np.pi * 100, 100000)
    x0 = (R - r) * np.cos(theta)
    y0 = (R - r) * np.sin(theta)
    x = x0 + (r - a) * np.cos(theta - R / r * theta) 
    y = y0 + (r - a) * np.sin(theta - R / r * theta) 
    plt.figure(figsize=(8, 8))
    plt.plot(x, y)    
````

````python
spirograph()
````

````python
spirograph(30, 53, -2)
````
